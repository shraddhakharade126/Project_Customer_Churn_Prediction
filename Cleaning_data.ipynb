```python
# import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset

ds=pd.read_csv("customer_churn_dataset.csv")
print(ds.to_string())
```

```python
ds = ds.drop(columns=['customerID','Dependents','tenure','InternetService','TotalCharges'])
ds
```

```python
ds['gender']=ds['gender'].str.replace("female","Female").str.replace("male","Male")
ds['Contract']=ds['Contract'].str.replace("month-to-Month","Month-to-month")
ds
```

```python
cols = ['Partner', 'PhoneService', 'MultipleLines','Churn']

for col in cols:
    ds[col] = ds[col].str.upper()
```

```python
# Numerical columns :impute with median
num_cols =['SeniorCitizen']
ds[num_cols]=ds[num_cols].fillna(ds[num_cols].median())

#categorical columns encode with mode
cat_col =['gender', 'Partner', 'PhoneService', 'MultipleLines','Contract', 'PaymentMethod', 'MonthlyCharges']
for col in cat_col:
    ds[col]=ds[col].fillna(ds[col].mode()[0])
```

```python
ds.isnull().sum()
```

```python
ds.dropna(subset=['Churn'],inplace=True)
```

```python
ds.info()
```

```python
ds['SeniorCitizen']=ds['SeniorCitizen'].astype(int)
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# List of columns to encode
cols_to_encode = ['gender', 'Partner', 'PhoneService', 'MultipleLines', 'Contract', 'PaymentMethod','Churn']

for col in cols_to_encode:
    # Fit the encoder on this column
    le.fit(ds[col])
    
    # Create mapping
    that_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print(f"Encoding Map for {col}: {that_mapping}")
    
    # Transform and overwrite the column
    ds[col] = le.transform(ds[col])
```
